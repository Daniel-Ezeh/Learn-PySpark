#### Loading CSV files

```python
from pyspark.sql import SparkSession

file_path = '/Users/nombauser/Desktop/GIT/MyGitRepos/Learn-PySpark/files/sample_1.csv'
json_file_path = '/Users/nombauser/Desktop/GIT/MyGitRepos/Learn-PySpark/files/sample_3.json'

spark = SparkSession.builder.appName("Loading files").getOrCreate()
```

```python
df = spark.read.format("csv") \
    .option("header", "true") \
    .option("inferSchema", "true")\
    .load(file_path)

df.show(4)

df.printSchema()
```

```python
df = spark.read.format("json") \
    .option("multiline", "true") \
    .load(json_file_path)
df.show()
df.printSchema()
```

```python
df.select(
    "boardgame",
    "credits.artists",
    "description",
    "game_stats"

).show(n=4, truncate=False)
```
